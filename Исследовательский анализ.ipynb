```python
number_empty_column = 0
for column in df.columns:
    if (len(df[df[column].isna()])) != 0:
        number_empty_column = number_empty_column + 1
        print(f'{number_empty_column} В столбце {column} - {len(df[df[column].isna()])} пропусков.')
```

**Примечание:** В таблице имеется 14 столбцов с пропусками.
* Можно сразу заполнить пропуски в столбце balcony_count значением 0, т.к. скорее всего балкон в квартире отсутствует, поэтому это значение не было заполнено.
* Пропуски в столбце is_apartment заполнить значением False, т.к. если значение не указано, то скорее всего это не апартаменты.
* В столбце floors_total всего 86 пропусков, при этом в столбце floor пропуски отсутствуют - пропуски в столбце floors_total можно заполнить значениями из столбца floor (этажей в доме точно не меньше, чем на котором находится квартира из объявления).

* Пропуски в столбцах living_area и kitchen_area заполним медианой, определенной в каждой категории квартир отдельно (в зависимости от количества комнат в квартире).

Остальные пропуски нужно рассмотреть внимательно.

```python
df.loc[df['balcony_count'].isna(), 'balcony_count'] = 0
df.loc[df['is_apartment'].isna(), 'is_apartment'] = False
df.loc[df['floors_total'].isna(), 'floors_total'] = df.loc[df['floors_total'].isna(), 'floor']
df.loc[df['locality_name'].isna(), 'locality_name'] = 'unknown'
df.loc[df['parks_around3000'].isna(), 'parks_around3000'] = 0
df.loc[df['ponds_around3000'].isna(), 'ponds_around3000'] = 0
```

Напишем функцию для заполнения пропусков в столбце  living_area:

```python
pd.options.mode.chained_assignment = None

unique_rooms_category = df['rooms_category'].unique()

empty_living_area = df[df['living_area'].isna()]
def fill_empty_living_area(column):
    for element in column:
        element_data = df[df['rooms_category'] == element]
        median = element_data['living_area'].median()
        empty_living_area.loc[empty_living_area['rooms_category'] == element, 'living_area'] = median
    return empty_living_area
        
fill_empty_living_area(unique_rooms_category)
df[df['living_area'].isna()] = empty_living_area
```

Напишем функцию для заполнения пропусков в столбце kitchen_area:

```python
empty_kitchen_area = df[df['kitchen_area'].isna()]
def fill_empty_kitchen_area(column):
    for element in column:
        element_data = df[df['rooms_category'] == element]
        median = element_data['kitchen_area'].median()
        empty_kitchen_area.loc[empty_kitchen_area['rooms_category'] == element, 'kitchen_area'] = median
    return empty_kitchen_area
        
fill_empty_kitchen_area(unique_rooms_category)
df[df['kitchen_area'].isna()] = empty_kitchen_area
```

### Добавим категорию этажа квартиры - первый, последний, другой:

Напишем функцию для присвоения категории этажу:

```python
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
```
